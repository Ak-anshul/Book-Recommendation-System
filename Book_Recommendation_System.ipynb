```python
num_ratings = book_ratings.groupby("Book-Title").count()["Book-Rating"].reset_index() 
```

```python
num_ratings.rename(columns={"Book-Rating":"NumOf_Ratings"},inplace=True)
num_ratings.head(20)
```

```python
avg_ratings = book_ratings.groupby("Book-Title")["Book-Rating"].mean().reset_index()
avg_ratings.rename(columns={"Book-Rating":"Avg_Ratings"},inplace=True)
```

```python
avg_ratings.head()
```

```python
dataset = pd.merge(num_ratings,avg_ratings,on="Book-Title")
```

```python
dataset_sorted = dataset.sort_values(by="Avg_Ratings",ascending=False)
dataset_sorted = dataset_sorted[dataset_sorted["NumOf_Ratings"]>=450]
dataset_sorted.shape
```

## Top 50 Popular Books

```python
dataset_sorted.head(50).reset_index()
```

## Popular Books on basis of Weighted Rating(Similiar to IMDB rating formula)

```python
c = dataset_sorted["Avg_Ratings"].mean()
def weighted_rating(x,c=c):
    v = x["NumOf_Ratings"]
    r = x["Avg_Ratings"]
    m = 450
    return ((v*r/(v+m))+(m*c/(v+m)))
```

```python
"""
r = Average rating for the book
v = Number of ratings for the book
m = Minimum number of ratings required to be listed in the chart(here ,450) 
c = Mean rating across all books
"""
```

```python
dataset_sorted["Weighted_Rating"] = dataset_sorted.apply(weighted_rating, axis=1)
```

```python
dataset_sorted.head() 
```

```python
books_sorted = dataset_sorted.sort_values(by="Weighted_Rating", ascending=False)
```

```python
books_sorted.head()
books_sorted.shape
```

```python
df = pd.merge(books_sorted,books,on="Book-Title")
df = df.drop_duplicates("Book-Title")
df.shape
```

```python
top_Books = df[["Book-Title","NumOf_Ratings","Avg_Ratings","Weighted_Rating","Book-Author","Year-Of-Publication","Publisher"]]
```

```python
top_Books.shape
```

```python
top_Books.head(50).reset_index()
```

## Hybrid (Recommendation based on weighted avg and avg rating (50% weight to both))

```python
# Standardisation of Avg and Weighting rating to bring them on same scale so we can Weighted score(50%) of both
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
top_Books[["Avg_Ratings","Weighted_Rating"]] = sc.fit_transform(top_Books[["Avg_Ratings","Weighted_Rating"]])
```

```python
top_Books.head()
```

```python
def function(data):
   return (data["Avg_Ratings"]*0.5 + data["Weighted_Rating"]*0.5) 
```

```python
top_Books["Score"] = top_Books.apply(function,axis = 1) 
```

```python
top_Books.sort_values(by="Score",ascending = False).reset_index()
```

## Collaborative Filtering Based Recommendation System

```python
UserWithRatings = pd.merge(ratings,books,on="ISBN")
```

```python
UserWithRatings.head()
```

```python
user_list = UserWithRatings.groupby("User-ID").count()["Book-Rating"]
user_list = user_list[user_list>=200]
user_list =user_list.index
## index is the 1st column 
```

```python
user_list
```

```python
UserWithRatings = UserWithRatings[UserWithRatings["User-ID"].isin(user_list)] 
```

```python
book_list = UserWithRatings.groupby("Book-Title")["Book-Rating"].count()
book_list.shape
```

```python
book_list = book_list[book_list>=50] 
book_list = book_list.index   
```

```python
final_df = UserWithRatings[UserWithRatings["Book-Title"].isin(book_list)]
```

```python
final_df.shape
```

```python
final_df.drop_duplicates()
```

```python
final_df = final_df.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
```

```python
final_df = final_df.fillna(0)
```

```python
final_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(final_df) 
```

```python
similarity.shape 
```

```python
similarity[0] 
```

```python
def recommend(book,n):
    if book not in final_df.index:
        print("This book "+ book +" is not in present in dataset")
    book_index = final_df.index.get_loc(book) 
    distance = similarity[book_index]
    book_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:int(n)+1]
    count = 1
    for i in book_list:
        print(str(count)+ " " +final_df.index[i[0]])
        count+=1
```

```python
print("Welcome To Book Recommendation System") 
book_name = input("Please enter the name of the Book ") 
n=input("How many similar Book do you need? ") 
recommend(book_name , n) 
```
